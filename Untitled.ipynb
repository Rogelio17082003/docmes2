```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import matplotlib.pyplot as plt

import seaborn as sns

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Dropout
data = pd.read_csv('mobile phone price prediction.csv')
data
```

```python
data= data.drop(['id'], axis=1)
```

# preparacion de los datos

```python
# los datos se rellenan con la moda  ya que son datos de tipo float pero no contienen decimal

data['Android_version'] = data['Android_version'].fillna(data['Android_version'].mode()[0])
data['Inbuilt_memory'] = data['Inbuilt_memory'].fillna(data['Inbuilt_memory'].mode()[0])
data['fast_charging'] = data['fast_charging'].fillna(data['fast_charging'].mode()[0])
data['Screen_resolution'] = data['Screen_resolution'].fillna(data['Screen_resolution'].mode()[0])
data['Processor'] = data['Processor'].fillna(data['Processor'].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

columns_to_encode = ['Name','Rating','Spec_score','No_of_sim','Ram','Battery','Display','Camera','External_Memory',	'Android_version',	'company'	,'Inbuilt_memory',	'fast_charging',	'Screen_resolution',	'Processor','Processor_name']  # lista de columnas a codificar

for column in columns_to_encode:
    data[column] = label_encoder.fit_transform(data[column])
```

```python
# conversion de atributos
data
```

```python
X = data.drop(['Price','Name'], axis=1)
Y = data['Price'].str.replace(',', '').astype(float)

X
xx = data[['Spec_score','Ram', 'External_Memory', 'company', 'Screen_resolution','Processor']]
```

```python
scaled = StandardScaler()
x_scaled = scaled.fit_transform(X)
x_scaled = pd.DataFrame(x_scaled, columns=X.columns)
x_scaled.head(5)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, Y, test_size=0.2, random_state=42)
x_train.head(5)
```

```python
modelo = LinearRegression() # Generar el modelo de LinearRegression

rfe = RFE(estimator=modelo, n_features_to_select=6, step=1)
rfe = rfe.fit(x_train, y_train)

for feature, selected in zip(X.columns, rfe.support_):
    print(f'{feature}: {selected}')

# Mostrar la clasificación de las características
print('Ranking de características:', rfe.ranking_)
print('Número óptimo de características :', rfe.n_features_)
```

```python
data.info()
```

```python
tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(x_train, y_train)

importancesFeactures = tree_model.feature_importances_

for feature, importanceData in zip(X.columns, importancesFeactures):
    print(f"Feacture: {feature}, Important:{importanceData}")
```

```python
x_train_selected = x_train[['Spec_score','Ram', 'External_Memory', 'company', 'Screen_resolution','Processor']]
x_test_selected = x_test[['Spec_score','Ram', 'External_Memory', 'company', 'Screen_resolution', 'Processor']]
x_train_selected.head(5)
```

```python
# Creando el modelo Keras
def create_model(inputShape):
    model = Sequential()
    model.add(Input(shape=(inputShape,)))  # Capa de entrada
    model.add(Dense(64, activation='relu'))
    model.add(Dense(32, activation='relu'))  # Capa oculta
    model.add(Dense(8, activation='relu'))  # Capa oculta
    model.add(Dense(1))  # Capa de salida
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

# Crear y entrenar los modelos
model = create_model(x_train_selected.shape[1])

# Ajustar modelo y registrar la historia de entrenamiento para el trazado posterior
history = model.fit(x_train_selected, y_train, validation_data=(x_test_selected, y_test), epochs=1000, verbose=0)

# Realizando predicciones con los datos de pruebam
y_pred = model.predict(x_test_selected)

# Calculando métricas para RFE
mae_RFE = mean_absolute_error(y_test, y_pred)
mse_RFE = mean_squared_error(y_test, y_pred)
rmse_RFE = np.sqrt(mse_RFE)
log_rmse_RFE = np.log(rmse_RFE)
r2_RFE = r2_score(y_test, y_pred)

# Imprimiendo métricas
print(f"RFE - MAE: {mae_RFE}, MSE: {mse_RFE}, RMSE: {rmse_RFE}, Log RMSE: {log_rmse_RFE}, R^2: {r2_RFE}")
```

```python
# Graficando Valores Reales vs Predicciones para RFE
plt.figure()
plt.scatter(y_test, y_pred, alpha=0.5)
lims = [np.min([plt.xlim(), plt.ylim()]),  np.max([plt.xlim(), plt.ylim()])]
plt.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
plt.xlabel('Valores Reales')
plt.ylabel('Predicciones')
plt.title('Valores Reales vs Predicciones RFE')
plt.show()

# Histograma de los residuos para RFE
plt.figure()
residuos_RFE = y_test - y_pred.flatten()
sns.histplot(residuos_RFE, kde=True)
plt.xlabel('Residuos')
plt.title('Distribución de los residuos Datos RFE')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_validate, KFold,cross_val_score
from sklearn.metrics import make_scorer, r2_score

# Crear y Entrenar el modelo RandomForestRegressor usando x_train_selected
forest_selected = RandomForestRegressor(random_state=42)
forest_selected.fit(x_train_selected, y_train)
y_pred_F = forest_selected.predict(x_test_selected)

#***************************************************+

# Calculando métricas para RandomForestRegressor con características RFE
mae_forest = mean_absolute_error(y_test, y_pred_F)
mse_forest = mean_squared_error(y_test, y_pred_F)
rmse_forest = np.sqrt(mse_forest)
log_rmse_forest = np.log(rmse_forest)
r2_forest = r2_score(y_test, y_pred_F)

# Imprimiendo métricas
print("\nMétricas para RandomForestRegressor con características RFE:")
print(f"MAE: {mae_forest}, MSE: {mse_forest}, RMSE: {rmse_forest}, Log RMSE: {log_rmse_forest}, R^2: {r2_forest}")
```

```python
# Graficando Valores Reales vs Predicciones de RandomForestRegressor con características RFE
plt.figure()
plt.scatter(y_test, y_pred_F)
lims = [np.min([plt.xlim(), plt.ylim()]),  np.max([plt.xlim(), plt.ylim()])]
plt.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
plt.xlabel('Valores Reales')
plt.ylabel('Predicciones')
plt.title('Valores Reales vs Predicciones RandomForestRegressor (RFE)')
plt.show()

# Histograma de los residuos para RandomForestRegressor con características RFE
plt.figure()
residuos_forestRFE = y_test - y_pred_F
sns.histplot(residuos_forestRFE, kde=True)
plt.xlabel('Residuos')
plt.title('Distribución de los residuos de RandomForestRegressor (RFE)')
plt.show()
```
